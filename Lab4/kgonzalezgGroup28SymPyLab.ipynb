## SymPyâ€™s Symbolic and Numercical Complex Evaluations

Captura web_1-4-2023_214111_www.wolframalpha.com.jpeg

```
from sympy import symbols, I, N, pi, expand
x1, y1, z1 = symbols("x1 y1 z1 ", real=True)  
x = 28 + I*x1
y = 30 + I*y1
z = 35 + I*z1

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))

w = N(1/(pi + I), 20)
print('w=',w)
```
